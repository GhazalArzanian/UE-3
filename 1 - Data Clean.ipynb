```python
import pandas as pd
```

```python
df = pd.read_csv("data/raw_global_youtube_statistics.csv", encoding="ISO-8859-1")
```

```python
df
```

```python
df.columns
```

```python
del df['Title']
del df['Abbreviation']
```

```python
cols=['rank', 'Youtuber', 'subscribers', 'video views', 'category',
       'uploads', 'Country', 'channel_type',
       'video_views_rank', 'country_rank',
       'highest_monthly_earnings',
       'created_year', 
       'Gross tertiary education enrollment (%)',
       'Unemployment rate']
```

```python
df_col= df[cols]
```

```python
df_col
```

```python
df_col.isna().sum()
```

```python
df_cleaned1=df_col.dropna(subset=["Country", "category", "country_rank", "created_year", "Unemployment rate"])
```

```python
df_cleaned1
```

```python
df_cleaned1["category"].value_counts()
```

```python
df_cleaned1.to_csv('data/cleaned_global_youtube_statistics.csv', index=False)
```

```python
df_cleaned1.columns
```

```python
df_country_cols=df_cleaned1[['subscribers', 'video views', 'uploads', 'channel_type', 'Country', 'highest_monthly_earnings', 
                            'Gross tertiary education enrollment (%)', 'Unemployment rate']]
```

```python
agg_funcs = {
    'subscribers': 'mean',
    'video views': 'mean',
    'uploads': 'mean',
    'highest_monthly_earnings': 'max',
    'Gross tertiary education enrollment (%)':'first',
    'Unemployment rate': 'first'
}
```

```python
df_countries=df_country_cols.groupby("Country", as_index=False).agg(agg_funcs)
```

```python
df_countries
```

```python
# Step 1: Count the total number of lines per country
country_counts = df_cleaned1['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Total']

# Step 2: Count the number of each category per country
category_counts = df_cleaned1.groupby(['Country', 'channel_type']).size().unstack(fill_value=0).reset_index()

# Step 3: Merge the total counts with the category counts
new_df = pd.merge(country_counts, category_counts, on='Country')

# Display the new DataFrame
new_df
```

```python
new_df.to_csv('data/cleaned_country_data.csv', index=False)
```

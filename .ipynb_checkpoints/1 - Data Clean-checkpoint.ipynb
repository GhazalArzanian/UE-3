```python
import chardet

# Olvasd be a fájl első néhány ezer byte-ját az encoding ellenőrzéséhez
with open('data/raw_global_youtube_statistics.csv', 'rb') as f:
    result = chardet.detect(f.read(10000))

print(result['encoding'])

# Használd a megadott karakterkódolást a CSV fájl beolvasásához
df = pd.read_csv('data/raw_global_youtube_statistics.csv', encoding=result['encoding'])

# Az adatok megjelenítése
print(df.head())
```

```python
df_col= df[cols]
```
